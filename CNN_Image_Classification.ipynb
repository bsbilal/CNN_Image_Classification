```python
import matplotlib.pyplot as plt

def plot_accuracy_and_loss(history):
    acc = history.history['acc']
    val_acc = history.history['val_acc']
    loss = history.history['loss']
    val_loss = history.history['val_loss']

    epochs = range(1, len(acc) + 1)

    plt.plot(epochs, acc, 'bo', label='Training acc')
    plt.plot(epochs, val_acc, 'b', label='Validation acc')
    plt.title('Training and validation accuracy')
    plt.legend()
    plt.show()

    plt.plot(epochs, loss, 'bo', label='Training loss')
    plt.plot(epochs, val_loss, 'b', label='Validation loss')
    plt.title('Training and validation loss')
    plt.legend()
    plt.show()
    

"çizim işlemlerine başlamak için matplotlib kütüphanesini import ediyoruz ve daha sonra sonuçları alabilmek için dizi içerisine bu değişkenleri tanımlıyoruz "
```

```python
from keras import layers, models, optimizers

model = models.Sequential()
model.add(layers.Conv2D(32, (3, 3), activation='relu', input_shape=(128,128, IMAGE_CHANNELS)))
model.add(layers.MaxPool2D(2, 2))
model.add(layers.Conv2D(64, (3, 3), activation='relu'))
model.add(layers.MaxPool2D(2, 2))
model.add(layers.Conv2D(128, (3, 3), activation='relu'))
model.add(layers.MaxPool2D(2, 2))
model.add(layers.Conv2D(128, (3, 3), activation='relu'))
model.add(layers.MaxPool2D(2, 2))
model.add(layers.Flatten())
model.add(layers.Dense(512, activation='relu'))
model.add(layers.Dense(1, activation='sigmoid'))




model.compile(loss='binary_crossentropy',
              optimizer=optimizers.RMSprop(lr=1e-4),
              metrics=['acc'])

model.summary()


"tanımladığımız değişkenler sayesinde 128x128 boyutlandırdığımız görselleri keras kütüphanesini import ederek çalıştırmaya başlatacağız ve aldığımız görsellerin özetini model.summary ile istiyoruz"
```

```python
from keras.preprocessing.image import ImageDataGenerator

train_datagen = ImageDataGenerator(rescale=1./255) 
test_datagen = ImageDataGenerator(rescale=1./255)

train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=(128, 128), 
    batch_size=10,
    class_mode='binary') 

validation_generator = test_datagen.flow_from_directory(
    validation_dir,
    target_size=(128, 128),
    batch_size=10,
    class_mode='binary')

"Görsellerimizin piksellerini işlemek için 0 ile 256 arası yenidenn boyutlandırıyoruz ve hedef boyutlarımızı tanımlıyoruz"
```

```python
history = model.fit_generator(
    train_generator,
    steps_per_epoch=50,
    epochs=30,
    validation_data=validation_generator,
    validation_steps=15)

"CNN MaxPool Egitimine berlirledigimiz degerlerle başlıyoruz"
```

```python
plot_accuracy_and_loss(history)

```

```python
test_generator = test_datagen.flow_from_directory(
    test_dir,
    target_size=(128, 128),
    batch_size=10,
    class_mode='binary')

test_loss, test_acc = model.evaluate_generator(test_generator, steps=30)
print('Test basarim orani:', test_acc)
```

```python
from keras.preprocessing.image import ImageDataGenerator

datagen = ImageDataGenerator(
    rotation_range=40,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest')

"Data augmentation metodunda fotoğrafı farklı açılara çevirip işlemler gerçekleştiriyoruz biz burada 40 derece açıyla çevirdik ve birtakım farklı görsel işlemler uyguladık"
```

```python
import random


from keras.preprocessing import image

fnames = [os.path.join(train_kedi_dir, fname) for fname in os.listdir(train_kedi_dir)]

img_path = fnames[random.randint(1,120)] 

img = image.load_img(img_path, target_size=(128, 128))

x = image.img_to_array(img) 

x = x.reshape((1,) + x.shape)

i = 0
for batch in datagen.flow(x, batch_size=1):
    plt.figure(i)
    imgplot = plt.imshow(image.array_to_img(batch[0]))
    i += 1
    if i % 4 == 0:
        break
plt.show()


"Bu aşamada ise random bir sayı aldık bunu eğitilecek fotoğrafımıza yaptığımız işlemleri kullanıcıya göstermek için yaptık diyebiliriz. Yüklediğimiz görseli işleyerek sistemimize uygun hale getirdik ve en son aşamada ona yaptığımız işlemleri gösterdik"
```

```python
model = models.Sequential()
model.add(layers.Conv2D(32, (3, 3), activation='relu', input_shape=(128, 128, 3)))
model.add(layers.MaxPool2D(2, 2))
model.add(layers.Conv2D(64, (3, 3), activation='relu'))
model.add(layers.MaxPool2D(2, 2))
model.add(layers.Conv2D(128, (3, 3), activation='relu'))
model.add(layers.MaxPool2D(2, 2))
model.add(layers.Conv2D(128, (3, 3), activation='relu'))
model.add(layers.MaxPool2D(2, 2))
model.add(layers.Flatten())
model.add(layers.Dense(512, activation='relu'))
model.add(layers.Dense(1, activation='sigmoid'))

model.compile(loss='binary_crossentropy',
              optimizer=optimizers.RMSprop(lr=1e-4),
              metrics=['acc'])

"Augmentation metodu sayesinde elimizdeki verilerin sayısını arttırmaya çalışıyoruz. Zaten görsellerimizi işlemiştik."
```

```python
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=40,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True)

test_datagen = ImageDataGenerator(rescale=1./255) 

train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=(128, 128),
    batch_size=10,
    class_mode='binary')

validation_generator = test_datagen.flow_from_directory(
    validation_dir,
    target_size=(128, 128),
    batch_size=10,
    class_mode='binary')

"Verisetimizi Augmentation için yeniden hazırlıyoruz bakalım şimdiki eğitim bize ne gösterecek.."
```

Bu da bize grafiğimize baktığımızda gördüğümüz şekilde düzensiz olmayan doğrular çizmiş durumda. İstediğimiz tam olarak bu değildi bazı şeyleri değiştirmemiz gerekiyor. Bir sonraki adıma geçelim: 

Augmentation'a Drop Out ekleyerek işlem yapma..


```python
model = models.Sequential()
model.add(layers.Conv2D(32, (3, 3), activation='relu', input_shape=(128, 128, 3)))
model.add(layers.MaxPool2D(2, 2))
model.add(layers.Conv2D(64, (3, 3), activation='relu'))
model.add(layers.MaxPool2D(2, 2))
model.add(layers.Conv2D(128, (3, 3), activation='relu'))
model.add(layers.MaxPool2D(2, 2))
model.add(layers.Conv2D(128, (3, 3), activation='relu'))
model.add(layers.MaxPool2D(2, 2))
model.add(layers.Flatten())
model.add(layers.Dropout(0.5))
model.add(layers.Dense(512, activation='relu'))
model.add(layers.Dense(1, activation='sigmoid'))

model.compile(loss='binary_crossentropy',
              optimizer=optimizers.RMSprop(lr=1e-4),
              metrics=['acc'])

"model.add(layers.Dropout(0.5)) satırı ile dropout eklemiş olduk"
```

```python
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=40,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True)

test_datagen = ImageDataGenerator(rescale=1./255) 

train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=(128, 128),
    batch_size=10,
    class_mode='binary')

validation_generator = test_datagen.flow_from_directory(
    validation_dir,
    target_size=(128, 128),
    batch_size=10,
    class_mode='binary')

"augmentation işleminde yaptığımız dönüştürme işleminin aynısını yapıyoruz."
```

```python
history = model.fit_generator(
    train_generator,
    steps_per_epoch=150,
    epochs=30,
    validation_data=validation_generator,
    validation_steps=45)
```

```python
plot_accuracy_and_loss(history)
```

```python
test_generator = test_datagen.flow_from_directory(
    test_dir,
    target_size=(128, 128),
    batch_size=10,
    class_mode='binary')

test_loss, test_acc = model.evaluate_generator(test_generator, steps=90)
print('Test başarım oranı:', test_acc)
```

# 3. Eğitim Yorumu :

Açıkcası istediğimiz sonuca ulaşmış gibi gözüküyoruz. Fakat gözüktüğü üzere belirli bir epochtan sonra verimiz mantıksız sonuçlar vermeye başladı. Burada kastettiğimiz mantıksız sonuç: "sürekli aynı seviyede seyretmek" olarak düşünülebilir. Maxpool ile overfit olan verilerimiz augmentation ve dropout metodlarının birleşimiyle çok iyi sonuç verdi denilebilir. Peki yazının başında bahsettiğimiz kırılma noktasını ayarlarsak nasıl bir sonuç elde ederiz? Beraber 4. aşamaya geçerek şu metodla bakalım : Epoch'u erken kesme deneyi

```python
model = models.Sequential()
model.add(layers.Conv2D(32, (3, 3), activation='relu', input_shape=(128, 128, 3)))
model.add(layers.MaxPool2D(2, 2))
model.add(layers.Conv2D(64, (3, 3), activation='relu'))
model.add(layers.MaxPool2D(2, 2))
model.add(layers.Conv2D(128, (3, 3), activation='relu'))
model.add(layers.MaxPool2D(2, 2))
model.add(layers.Conv2D(128, (3, 3), activation='relu'))
model.add(layers.MaxPool2D(2, 2))
model.add(layers.Flatten())
model.add(layers.Dropout(0.5))
model.add(layers.Dense(512, activation='relu'))
model.add(layers.Dense(1, activation='sigmoid'))

model.compile(loss='binary_crossentropy',
              optimizer=optimizers.RMSprop(lr=1e-4),
              metrics=['acc'])

"Dropout metoduyla aynı şekilde verilerimizi ayarlıyoruz"
```

```python
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=40,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True)

test_datagen = ImageDataGenerator(rescale=1./255) 

train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=(128, 128),
    batch_size=10,
    class_mode='binary')

validation_generator = test_datagen.flow_from_directory(
    validation_dir,
    target_size=(128, 128),
    batch_size=10,
    class_mode='binary')
```

```python
print("Erken epoch kesme ayarımız için normalde 30 olan epoch sayımızı 5'de sınırlandırdık.")
history = model.fit_generator(
    train_generator,
    steps_per_epoch=150,
    epochs=5, 
    validation_data=validation_generator,
    validation_steps=45) 
```

```python
plot_accuracy_and_loss(history)
```

```python
test_generator = test_datagen.flow_from_directory(
    test_dir,
    target_size=(128,128),
    batch_size=10,
    class_mode='binary')

test_loss, test_acc = model.evaluate_generator(test_generator, steps=90)
print('Test başarım oranı:', test_acc)
```
