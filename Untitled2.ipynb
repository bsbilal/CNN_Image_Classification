```python
dataset_dir='input/image_dataset'
train_dir = os.path.join(base_dir, 'train')
validation_dir = os.path.join(base_dir, 'validation')
test_dir = os.path.join(base_dir, 'test')
```

```python
train_kedi_dir = os.path.join(train_dir, 'kedi')
train_aycicegi_dir = os.path.join(train_dir, 'aycicegi')

validation_kedi_dir = os.path.join(validation_dir, 'kedi')
validation_aycicegi_dir = os.path.join(validation_dir, 'aycicegi')

test_kedi_dir = os.path.join(test_dir, 'kedi')
test_aycicegi_dir = os.path.join(test_dir, 'aycicegi')
```

```python
fnames = ['kedi.{}.jpg'.format(i) for i in range(150)]
for fname in fnames:
    src = os.path.join(dataset_dir, fname)
    dst = os.path.join(train_kedi_dir, fname)
    shutil.copyfile(src, dst)

fnames = ['kedi.{}.jpg'.format(i) for i in range(150, 190)]
for fname in fnames:
    src = os.path.join(dataset_dir, fname)
    dst = os.path.join(validation_kedi_dir, fname)
    shutil.copyfile(src, dst)
    
fnames = ['kedi.{}.jpg'.format(i) for i in range(190, 200)]
for fname in fnames:
    src = os.path.join(dataset_dir, fname)
    dst = os.path.join(test_kedi_dir, fname)
    shutil.copyfile(src, dst)
    
fnames = ['aycicegi.{}.jpg'.format(i) for i in range(150)]
for fname in fnames:
    src = os.path.join(dataset_dir, fname)
    dst = os.path.join(train_aycicegi_dir, fname)
    shutil.copyfile(src, dst)

fnames = ['aycicegi.{}.jpg'.format(i) for i in range(150, 190)]
for fname in fnames:
    src = os.path.join(dataset_dir, fname)
    dst = os.path.join(validation_aycicegi_dir, fname)
    shutil.copyfile(src, dst)
    
fnames = ['aycicegi.{}.jpg'.format(i) for i in range(190, 200)]
for fname in fnames:
    src = os.path.join(dataset_dir, fname)
    dst = os.path.join(test_aycicegi_dir, fname)
    shutil.copyfile(src, dst)
```

```python
import matplotlib.pyplot as plt

def plot_accuracy_and_loss(history):
    acc = history.history['acc']
    val_acc = history.history['val_acc']
    loss = history.history['loss']
    val_loss = history.history['val_loss']

    epochs = range(1, len(acc) + 1)

    plt.plot(epochs, acc, 'bo', label='Training acc')
    plt.plot(epochs, val_acc, 'b', label='Validation acc')
    plt.title('Training and validation accuracy')
    plt.legend()
    plt.show()

    plt.plot(epochs, loss, 'bo', label='Training loss')
    plt.plot(epochs, val_loss, 'b', label='Validation loss')
    plt.title('Training and validation loss')
    plt.legend()
    plt.show()
```

```python
from keras import layers, models, optimizers

model = models.Sequential()
model.add(layers.Conv2D(32, (3, 3), activation='relu', input_shape=(128, 128, 3)))
model.add(layers.MaxPool2D(2, 2))
model.add(layers.Conv2D(64, (3, 3), activation='relu'))
model.add(layers.MaxPool2D(2, 2))
model.add(layers.Conv2D(128, (3, 3), activation='relu'))
model.add(layers.MaxPool2D(2, 2))
model.add(layers.Conv2D(128, (3, 3), activation='relu'))
model.add(layers.MaxPool2D(2, 2))
model.add(layers.Flatten())
model.add(layers.Dense(512, activation='relu'))
model.add(layers.Dense(1, activation='sigmoid'))




model.compile(loss='binary_crossentropy',
              optimizer=optimizers.RMSprop(lr=1e-4),
              metrics=['acc'])

model.summary()
```

```python
from keras.preprocessing.image import ImageDataGenerator

# Piksel değerlerini [0, 1] ile [0, 255] arasında yeniden ölçeklendirir
train_datagen = ImageDataGenerator(rescale=1./255) 
test_datagen = ImageDataGenerator(rescale=1./255)

train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=(128, 128), # train resimlerini 224 x 224 piksel değerinde boyutlandırıyoruz.
    batch_size=10,
    class_mode='binary') 

validation_generator = test_datagen.flow_from_directory(
    validation_dir,
    target_size=(128, 128), # validation resimlerini 224 x 224 piksel değerinde boyutlandırıyoruz.
    batch_size=10,
    class_mode='binary')
```

```python
history = model.fit_generator(
    train_generator,
    steps_per_epoch=100,
    epochs=30,
    validation_data=validation_generator,
    validation_steps=50)
```

```python
plot_accuracy_and_loss(history)

```

```python
test_generator = test_datagen.flow_from_directory(
    test_dir,
    target_size=(224, 224),
    batch_size=16,
    class_mode='binary')

test_loss, test_acc = model.evaluate_generator(test_generator, steps=90)
print('test acc:', test_acc)
```
